```python
import base64
import numpy as np
import warnings
warnings.filterwarnings("ignore")

import io
from PIL import Image
from src.model import ReceiptOCR_DefaultModel
from src.engine import ReceiptOCR_DefaultEngine
```

## Load model network and weight

```python
detector_cfg = 'configs/craft_config.yaml'
recognizer_cfg = 'configs/star_config.yaml'

model = ReceiptOCR_DefaultModel(detector_cfg, recognizer_cfg)
```

## Load Engine

```python
engine = ReceiptOCR_DefaultEngine(model)
```

## Input Data Preparation

```python
image_file = 'data/tes.jpg'

with open(image_file, "rb") as f:
    im_bytes = f.read()

im_b64 = base64.b64encode(im_bytes).decode('utf-8')
```

```python
img_bytes = base64.b64decode(im_b64.encode('utf-8'))
image = Image.open(io.BytesIO(img_bytes))
image = np.array(image)
```

## Prediction

```python
engine.predict(image)
```
